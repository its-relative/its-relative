```python
# For the categorical variables, we can use:

# Bar plot to visualize the frequency of each category in each categorical variable.
# Box plot to visualize the distribution of each category in each categorical variable.
# Violin plot to visualize the distribution of each category in each categorical variable.

import seaborn as sns

# create a new DataFrame with only the categorical variables
cat_df = df[['Seat comfort', 'Departure/Arrival time convenient', 'Food and drink', 'Gate location', 
             'Inflight wifi service', 'Inflight entertainment', 'Online support', 'Ease of Online booking', 
             'On-board service', 'Leg room service', 'Baggage handling', 'Checkin service', 
             'Cleanliness', 'Online boarding']]

# create a grid of count plots for each categorical variable
g = sns.catplot(data=cat_df.melt(), x='variable', kind='count', col='value', col_wrap=3)

# Give the x-labels some rotation so that they are visible
g.set_xticklabels(rotation=45, horizontalalignment='right')

# set the titles for each subplot
g.set_titles('{col_name}={col_var}')

# adjust the spacing between the subplots
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Frequency of Categorical Variables', fontsize=16)

plt.show()
```
