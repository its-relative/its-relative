```
def checknumber(n):
  if (n == 0 or n == 1 or n == 2):
    return "true"

  a = 0
  b = 1
  while b< n:
    c = a+b
    a = b
    b = c
  if (b == n):
    return "true"
  return "false"

n = int(input())
print(checknumber(n))
```
