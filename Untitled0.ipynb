```
def checkPalindrome(num):
  x = str(num)
  y = str(num)[::-1]
  if x == y:
    return True
  else:
    return False 
		
num = int(input())
isPalindrome = checkPalindrome(num)
if(isPalindrome):
	print('true')
else:
	print('false')
```
